```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LSTM, GlobalMaxPooling1D, Embedding, Bidirectional, TimeDistributed
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from preprocess_text import remove_punctuations
```

```python
# Load Dataset and reemove punctuations
data = pd.read_csv("data/orig_key.csv")
data
```

```python
X = data["DATA"]
Y = data["LABEL"]
# SENT = data.groupby("SENT")
# sent = []
# master_sent = []
# for name,group in SENT:
#     print(group)
```

```python
x = np.array([np.array(item) for item in X.values])
y = np.array([np.array(item) for item in Y.values])
x
y
```

```python
all_data = []
all_words = []
for key,label,sent in data.values:
    for words in key.split(" "):
        if words is not "":
            all_data.append((words,label))
            all_words.append(words)
print(all_data)
all_words = remove_punctuations(all_words)
print(all_words)
```

```python
all_pairs = []
idx = 0
while idx < len(all_data):
    all_pairs.append(all_data[idx:idx+3])
    idx += 3
```

```python
all_pairs
```

```python
x = []
y = []
for d in all_pairs:
    x.append(d[0][0] + " " + d[1][0] + " " + d[2][0])
    y.append(str(d[0][1]) + " " + d[1][1] + " " + d[2][1])
x = np.array(x)
print(x)
y = np.array(y)
print(y)
```

```python
VOCAB_LENGTH = 20000
tokenizer = Tokenizer(num_words=VOCAB_LENGTH,oov_token="UNK")
tokenizer.fit_on_texts(all_words)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33)
print(x_train.shape)
print(y_train.shape)
```

```python
sequence_train = tokenizer.texts_to_sequences(x_train)
print(sequence_train)
sequence_test = tokenizer.texts_to_sequences(x_test)
print(sequence_test)
```

```python
label_tokenizer = Tokenizer(num_words=VOCAB_LENGTH)
label_tokenizer.fit_on_texts(y_train)
label_seq_train = label_tokenizer.texts_to_sequences(y_train)
label_seq_test = label_tokenizer.texts_to_sequences(y_test)
```

```python
label_tokenizer.word_index
```

```python
label_seq_train
```

```python
word2idx = tokenizer.word_index
print(word2idx)
V = len(word2idx)
print(V)
```

```python
T = 10
x_train = pad_sequences(sequence_train,maxlen=T,padding='post')
print(sequence_test)
T = x_train.shape[1]
print(f"max length is : {T}")
x_test = pad_sequences(sequence_test,maxlen=T,padding='post')
x_train.shape
```

```python
M = 50
K = 3
D = 50
i = Input(shape=(T,))
x = Embedding(V + 1, D)(i)
# x = Bidirectional(tf.keras.layers.LSTM(M))(x)
x = LSTM(M, return_sequences=True)(x)
x = GlobalMaxPooling1D()(x)
#x = TimeDistributed(Dense(128, activation='relu'))(x)
x = Dense(K, activation='sigmoid')(x)

model = Model(i, x)
model.compile(loss='categorical_crossentropy',
              optimizer='adam', metrics=["accuracy"])
model.summary()
```

```python
print(sequence_train)
print(type(sequence_train))
print(x_train)
print(x_train.shape)
print(type(y_train))
print(type(label_seq_train))
print((label_seq_train))

l_seq = [np.array(item) for item in label_seq_train]
print(l_seq)
print(type(l_seq))
```

```python
y_train = pad_sequences(label_seq_train,maxlen=3, padding='post')
y_test = pad_sequences(label_seq_test,maxlen=3, padding='post')
print(y_train.shape)
y_train = np.expand_dims(y_train,2)
y_train.shape
```

```python
x_train[0]
```

```python
r = model.fit(x_train, y_train, validation_data=(
    x_test, y_test), epochs=300)
```

```python
plt.plot(r.history["loss"], label="loss")
plt.plot(r.history["val_loss"], label="val_loss")
plt.legend()
plt.show()
```

```python
plt.plot(r.history["accuracy"], label="accuracy")
plt.plot(r.history["val_accuracy"], label="val_accuracy")
plt.legend()
plt.show()
```
